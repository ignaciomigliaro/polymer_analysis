```python
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
import numpy as np 
from scipy.stats import gaussian_kde


def make_filename(c, d, solvent):
    """
    Build filename depending on solvent mode.
    solvent = "odcb", "pdcb", or "both"
    """
    if solvent == "odcb":
        return f"sasa_c{c}_{d}_odcb.xvg"
    elif solvent == "pdcb":
        return f"sasa_c{c}_{d}_pdcb.xvg"
    else:  # "both"
        # try: _odcb â†’ _pdcb â†’ no suffix
        return [
            f"sasa_c{c}_{d}_odcb.xvg",
            f"sasa_c{c}_{d}_pdcb.xvg",
            f"sasa_c{c}_{d}.xvg",
        ]

def parse_xvg(file_path):
    """Safe parser that avoids iterator exhaustion and uneven columns."""
    time, sasa = [], []
    with open(file_path) as f:
        for line in f:
            if line.startswith(("#", "@")) or not line.strip():
                continue
            parts = line.split()
            if len(parts) >= 2:
                try:
                    time.append(float(parts[0]))
                    sasa.append(float(parts[1]))
                except ValueError:
                    continue
    return time, sasa


def plot_sasa(carbons, degrees, base_dir=".", solvent="both",
              title="SASA over Time", xlabel="Time (ps)", ylabel="SASA (nmÂ²)",
              output_file="sasa_plot.png"):

    base_dir = Path(base_dir)
    plt.figure(figsize=(16, 10))

    # --- Count total traces expected per solvent for palette size ---
    N = len(carbons) * len(degrees)

    palettes = {
        "odcb": sns.color_palette("Blues", N),
        "pdcb": sns.color_palette("Reds", N),
        "raw":  sns.color_palette("Greens", N),
    }

    color_index = {"odcb": 0, "pdcb": 0, "raw": 0}

    for c in carbons:
        for d in degrees:

            filenames = make_filename(c, d, solvent)
            if isinstance(filenames, str):
                filenames = [filenames]

            for fname in filenames:

                file_path = base_dir / fname
                if not file_path.exists():
                    continue

                data = parse_xvg(file_path)
                if not data:
                    print(f"âŒ No valid data in {file_path}")
                    continue

                time, sasa = data[0], data[1]

                # --- Determine solvent grouping for color ---
                suffix = fname.replace(".xvg", "").split("_")[-1]
                if suffix.isdigit():     # means no suffix â†’ "raw"
                    solvent_group = "raw"
                else:
                    solvent_group = suffix  # 'odcb' or 'pdcb'

                # --- Assign next unique color from the solvent palette ---
                color = palettes[solvent_group][color_index[solvent_group]]
                color_index[solvent_group] += 1

                label = f"C{c} - {d}% ({solvent_group})"

                plt.plot(time, sasa, label=label, linewidth=2, color=color)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="System", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.savefig(output_file, dpi=600)
    print(f"âœ… Plot saved: {output_file}")
    plt.show()



def plot_kde_sasa(carbons, degrees, base_dir=".", solvent="both",
                  title="SASA Distribution (KDE)", xlabel="SASA (nmÂ²)",
                  ylabel="Density", output_file="sasa_kde.png"):
    """
    KDE distributions of SASA for multiple carbon lengths, degrees, and solvents.
    Uses distinct color palettes per solvent to avoid overlap.

    Solvents:
        odcb â†’ blue palette
        pdcb â†’ red palette
        raw  â†’ green palette
    """

    def make_filenames(c, d, solvent):
        if solvent == "odcb":
            return [f"sasa_c{c}_{d}_odcb.xvg"]
        elif solvent == "pdcb":
            return [f"sasa_c{c}_{d}_pdcb.xvg"]
        else:
            return [
                f"sasa_c{c}_{d}_odcb.xvg",
                f"sasa_c{c}_{d}_pdcb.xvg",
                f"sasa_c{c}_{d}.xvg",
            ]

    base_dir = Path(base_dir)
    plt.figure(figsize=(8, 5))

    # Count possible lines per solvent group for palette sizing
    N = len(carbons) * len(degrees)

    palettes = {
        "odcb": sns.color_palette("Blues", N),
        "pdcb": sns.color_palette("Reds", N),
        "raw":  sns.color_palette("Greens", N),
    }

    color_index = {"odcb": 0, "pdcb": 0, "raw": 0}

    for c in carbons:
        for d in degrees:
            filenames = make_filenames(c, d, solvent)

            for fname in filenames:
                file_path = base_dir / fname
                if not file_path.exists():
                    continue

                data = parse_xvg(file_path)
                if not data:
                    print(f"âŒ No valid data in {file_path}")
                    continue

                sasa_values = data[1]

                # Label solvent correctly
                suffix = fname.replace(".xvg", "").split("_")[-1]
                if suffix.isdigit():
                    solvent_group = "raw"
                else:
                    solvent_group = suffix  # odcb or pdcb

                # Assign next color from that palette
                color = palettes[solvent_group][color_index[solvent_group]]
                color_index[solvent_group] += 1

                label = f"C{c} - {d}% ({solvent_group})"

                sns.kdeplot(sasa_values, label=label, linewidth=2, color=color)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="System", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(output_file, dpi=600)
    print(f"âœ… KDE plot saved as: {output_file}")
    plt.show()

    
def plot_avg_delta_sasa(carbons, degrees, base_dir=".", solvent="both",
                        title="Average Î”SASA vs Carbon Length",
                        xlabel="Carbon Chain Length (Câ‚™)",
                        ylabel="Average Î”SASA (nmÂ²)",
                        output_file="avg_delta_sasa_line.png"):
    """
    Compute & plot average Î”SASA for each carbon length and degree
    for the solvents oDCB and pDCB.

    Solvent options:
        "odcb" â†’ only *_odcb.xvg
        "pdcb" â†’ only *_pdcb.xvg
        "both" â†’ plot both solvents
    """

    base_dir = Path(base_dir)
    plt.figure(figsize=(8, 5))

    # --- Color palettes for the two solvents ---
    palettes = {
        "odcb": sns.color_palette("Blues", len(degrees)),
        "pdcb": sns.color_palette("Reds", len(degrees)),
    }

    # Determine which solvents to plot
    solvent_order = []
    if solvent in ("odcb", "both"):
        solvent_order.append("odcb")
    if solvent in ("pdcb", "both"):
        solvent_order.append("pdcb")

    # Loop 1 solvent at a time â†’ ensures both plot separately
    for solvent_group in solvent_order:

        palette = palettes[solvent_group]

        # One line per degree
        for i, d in enumerate(degrees):
            avg_deltas = []

            for c in carbons:

                # Build filename for THIS solvent
                if solvent_group == "odcb":
                    fname = f"sasa_c{c}_{d}_odcb.xvg"
                else:
                    fname = f"sasa_c{c}_{d}_pdcb.xvg"

                file_path = base_dir / fname

                if not file_path.exists():
                    avg_deltas.append(np.nan)
                    continue

                data = parse_xvg(file_path)
                if not data:
                    avg_deltas.append(np.nan)
                    continue

                sasa = np.array(data[1])
                delta_sasa = sasa - sasa.mean()
                avg_deltas.append(np.mean(np.abs(delta_sasa)))

            # Plot line FOR THIS solvent and degree
            label = f"{d}% ({solvent_group})"
            color = palette[i]

            plt.plot(carbons, avg_deltas, marker="o",
                     linewidth=2, color=color, label=label)

    # Styling
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="Functionalization + Solvent",
               bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(output_file, dpi=600)

    print(f"âœ… Line plot saved as: {output_file}")
    plt.show()


def plot_sasa_per_carbon(carbons, degrees, base_dir=".", solvent="both",
                         title="Average SASA per Carbon",
                         xlabel="Carbon Chain Length (Câ‚™)",
                         ylabel="SASA per Carbon (nmÂ²/atom)",
                         output_file="sasa_per_carbon.png"):
    """
    Plot mean SASA per carbon for multiple chain lengths and functionalization degrees,
    with full solvent awareness.
    
    Solvent options:
        "odcb" â†’ use *_odcb.xvg
        "pdcb" â†’ use *_pdcb.xvg
        "both" â†’ plot both solvents separately
    """

    base_dir = Path(base_dir)
    plt.figure(figsize=(8, 5))

    # Color palettes for each solvent group
    palettes = {
        "odcb": sns.color_palette("Blues", len(degrees)),
        "pdcb": sns.color_palette("Reds", len(degrees)),
    }

    # Determine which solvents to plot
    solvent_order = []
    if solvent in ("odcb", "both"):
        solvent_order.append("odcb")
    if solvent in ("pdcb", "both"):
        solvent_order.append("pdcb")

    # Loop over each solvent first
    for solvent_group in solvent_order:

        palette = palettes[solvent_group]

        # One line per degree for this solvent
        for i, d in enumerate(degrees):
            mean_sasa_per_c = []

            for c in carbons:
                # Build filename for this solvent
                if solvent_group == "odcb":
                    fname = f"sasa_c{c}_{d}_odcb.xvg"
                else:
                    fname = f"sasa_c{c}_{d}_pdcb.xvg"

                fpath = base_dir / fname

                if not fpath.exists():
                    print(f"âš ï¸ Missing file: {fpath}")
                    mean_sasa_per_c.append(np.nan)
                    continue

                data = parse_xvg(fpath)
                if not data:
                    print(f"âŒ No valid data in {fpath}")
                    mean_sasa_per_c.append(np.nan)
                    continue

                sasa = np.array(data[1])
                mean_sasa_per_c.append(sasa.mean() / c)

            # Plot one line per (degree Ã— solvent)
            label = f"{d}% ({solvent_group})"
            color = palette[i]

            plt.plot(carbons, mean_sasa_per_c, marker="o",
                     linewidth=2, color=color, label=label)

    # Styling
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="Functionalization + Solvent",
               bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(output_file, dpi=600)
    print(f"âœ… Plot saved as: {output_file}")
    plt.show()


def plot_normalized_sasa(carbons, degrees, base_dir=".", solvent="both",
                         title="Normalized Mean SASA per Polymer",
                         xlabel="Carbon Chain Length (Câ‚™)",
                         ylabel="âŸ¨SASA / âŸ¨SASAâŸ©âŸ© (unitless)",
                         output_file="normalized_mean_sasa.png"):
    """
    Compute and plot the mean normalized SASA (SASA / mean(SASA)) for each polymer,
    separated by solvent (oDCB or pDCB). One line per (degree Ã— solvent).

    Filenames:
        sasa_c{carbon}_{degree}_odcb.xvg
        sasa_c{carbon}_{degree}_pdcb.xvg
    """

    base_dir = Path(base_dir)
    plt.figure(figsize=(8, 5))

    # Solvent-specific color palettes
    palettes = {
        "odcb": sns.color_palette("Blues", len(degrees)),
        "pdcb": sns.color_palette("Reds", len(degrees)),
    }

    # Determine which solvents to include
    solvent_order = []
    if solvent in ("odcb", "both"):
        solvent_order.append("odcb")
    if solvent in ("pdcb", "both"):
        solvent_order.append("pdcb")

    # Loop by solvent first â†’ ensures both plot separately
    for solvent_group in solvent_order:

        palette = palettes[solvent_group]

        for i, d in enumerate(degrees):
            normalized_means = []

            for c in carbons:
                if solvent_group == "odcb":
                    fname = f"sasa_c{c}_{d}_odcb.xvg"
                else:
                    fname = f"sasa_c{c}_{d}_pdcb.xvg"

                file_path = base_dir / fname

                if not file_path.exists():
                    print(f"âš ï¸ Missing file: {file_path}")
                    normalized_means.append(np.nan)
                    continue

                data = parse_xvg(file_path)
                if not data:
                    print(f"âŒ No valid data in {file_path}")
                    normalized_means.append(np.nan)
                    continue

                sasa = np.array(data[1])
                norm_sasa = sasa / sasa.mean()
                normalized_means.append(norm_sasa.mean())

            # Plot one line per degree Ã— solvent
            label = f"{d}% ({solvent_group})"
            color = palette[i]

            plt.plot(carbons, normalized_means, marker='o', linewidth=2,
                     color=color, label=label)

    # Style
    plt.axhline(1.0, color='gray', linestyle='--', linewidth=1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="Functionalization + Solvent",
               bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(output_file, dpi=600)

    print(f"âœ… Normalized mean SASA plot saved as: {output_file}")
    plt.show()

    
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

def parse_xvg(file_path):
    """Parse an .xvg file and return numerical data as a list of lists."""
    with open(file_path, "r") as file:
        lines = [l for l in file if not (l.startswith("#") or l.startswith("@"))]
    data = [list(map(float, l.split())) for l in lines if l.strip()]
    return list(zip(*data)) if data else []


def plot_rmsd_normalized_sasa(carbons, degrees, base_dir=".", solvent="both",
                              title="RMSD of Normalized SASA per Polymer",
                              xlabel="Carbon Chain Length (Câ‚™)",
                              ylabel="RMSD[ SASA / âŸ¨SASAâŸ© ] (unitless)",
                              output_file="rmsd_normalized_sasa.png"):
    """
    Compute and plot RMSD of normalized SASA for each carbon length and degree,
    separated by solvent. Produces one line per (degree Ã— solvent).

    Filenames:
        sasa_c{carbon}_{degree}_odcb.xvg
        sasa_c{carbon}_{degree}_pdcb.xvg
    """

    base_dir = Path(base_dir)
    plt.figure(figsize=(8, 5))

    # Color palettes for each solvent type
    palettes = {
        "odcb": sns.color_palette("Blues", len(degrees)),
        "pdcb": sns.color_palette("Reds", len(degrees)),
    }

    # Select which solvents to include
    solvent_order = []
    if solvent in ("odcb", "both"):
        solvent_order.append("odcb")
    if solvent in ("pdcb", "both"):
        solvent_order.append("pdcb")

    # Loop by solvent to guarantee both plot independently
    for solvent_group in solvent_order:

        palette = palettes[solvent_group]

        for i, d in enumerate(degrees):
            rmsd_values = []

            for c in carbons:
                # Build filename for THIS solvent
                if solvent_group == "odcb":
                    fname = f"sasa_c{c}_{d}_odcb.xvg"
                else:
                    fname = f"sasa_c{c}_{d}_pdcb.xvg"

                file_path = base_dir / fname

                if not file_path.exists():
                    print(f"âš ï¸ Missing file: {file_path}")
                    rmsd_values.append(np.nan)
                    continue

                data = parse_xvg(file_path)
                if not data:
                    print(f"âŒ No valid data in {file_path}")
                    rmsd_values.append(np.nan)
                    continue

                sasa = np.array(data[1])
                norm_sasa = sasa / sasa.mean()
                rmsd = np.sqrt(np.mean((norm_sasa - 1.0) ** 2))
                rmsd_values.append(rmsd)

            # Plot the line for this (degree Ã— solvent)
            label = f"{d}% ({solvent_group})"
            color = palette[i]

            plt.plot(carbons, rmsd_values, marker='o', linewidth=2,
                     color=color, label=label)

    # Formatting
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title="Functionalization + Solvent",
               bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.savefig(output_file, dpi=600)

    print(f"âœ… RMSD[ SASA / âŸ¨SASAâŸ© ] plot saved as: {output_file}")
    plt.show()
```

### End-to-End distance

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.lib.distances import calc_bonds

def plot_end_to_end_multiple(
        basepaths,
        atom1,
        atom2,
        labels=None,
        output_prefix="end_to_end_overlay"
    ):
    """
    Compute and overlay end-to-end distances for multiple trajectories.
    
    Parameters
    ----------
    basepaths : list of str
        Each entry is the base path without extension, e.g. "/path/to/prod/prod"
        The function will load {path}.tpr and {path}.xtc.
    atom1, atom2 : int
        GROMACS atom indices of the two end atoms (1-based).
    labels : list of str, optional
        Names for each line in the plot. If None, basepaths are used.
    output_prefix : str
        Prefix for output files (.png and individual .dat files).
    """

    # Convert GROMACS â†’ MDAnalysis indices
    idx1 = atom1 - 1
    idx2 = atom2 - 1

    if labels is None:
        labels = [f"traj_{i+1}" for i in range(len(basepaths))]

    plt.figure(figsize=(10, 5))

    for basepath, label in zip(basepaths, labels):
        tpr_path = basepath + ".tpr"
        xtc_path = basepath + ".xtc"

        print(f"\nLoading: {basepath}")
        print(f"  TPR: {tpr_path}")
        print(f"  XTC: {xtc_path}")

        u = mda.Universe(tpr_path, xtc_path)

        atom_start = u.atoms[idx1]
        atom_end   = u.atoms[idx2]

        distances = []
        times = []

        # Compute PBC-corrected end-to-end distance
        for ts in u.trajectory:
            d = calc_bonds(
                atom_start.position[np.newaxis, :],
                atom_end.position[np.newaxis, :],
                box=ts.dimensions
            )[0]
            distances.append(d)
            times.append(ts.time)

        distances = np.array(distances)
        times = np.array(times)

        # Save data file for this trajectory
        dat_file = f"{output_prefix}_{label}.dat"
        np.savetxt(dat_file, np.column_stack((times, distances)),
                   header="time(ps) distance(nm)")
        print(f"  Saved data -> {dat_file}")

        # Plot overlay
        plt.plot(times, distances, lw=1.3, label=label)

    # Finalize plot
    plt.xlabel("Time (ps)", fontsize=14)
    plt.ylabel("End-to-End Distance (nm)", fontsize=14)
    plt.title("End-to-End Distance vs Time (PBC-corrected)", fontsize=16)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()

    png_file = f"{output_prefix}.png"
    plt.savefig(png_file, dpi=300)
    plt.show()
    print(f"\nSaved combined plot -> {png_file}")
```

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.lib.distances import calc_bonds
from scipy.stats import gaussian_kde
import os

def kde_end_to_end_multiple(
        basepaths,
        atom1,
        atom2,
        labels=None,
        output_prefix="end_to_end_kde",
        bandwidth=None
    ):
    """
    Computes KDE curves of end-to-end distance for multiple trajectories.
    Reuses cached .dat files when present.

    Parameters
    ----------
    basepaths : list of str
        Base paths without extension, e.g., "/path/to/prod/prod".
        The function loads basepath.tpr and basepath.xtc.
    atom1, atom2 : int
        GROMACS atom indices (1-based).
    labels : list of str
        Names for each KDE curve.
    output_prefix : str
        Prefix for saved files (.dat and .png).
    bandwidth : float or None
        Optional bandwidth for gaussian_kde.
    """

    idx1 = atom1 - 1
    idx2 = atom2 - 1

    if labels is None:
        labels = [f"traj_{i+1}" for i in range(len(basepaths))]

    plt.figure(figsize=(10, 5))

    for basepath, label in zip(basepaths, labels):

        dat_file = f"{output_prefix}_{label}.dat"

        # =======================================================
        # 1. USE EXISTING .dat IF PRESENT
        # =======================================================
        if os.path.exists(dat_file):
            print(f"\nFound cached data â†’ {dat_file}")
            dists = np.loadtxt(dat_file)[:, 1]  # load only distance column

        else:
            # =======================================================
            # 2. OTHERWISE COMPUTE FROM TRAJECTORY
            # =======================================================
            print(f"\nComputing distances for {label}")

            tpr_path = basepath + ".tpr"
            xtc_path = basepath + ".xtc"

            print(f"  TPR: {tpr_path}")
            print(f"  XTC: {xtc_path}")

            u = mda.Universe(tpr_path, xtc_path)

            atom_start = u.atoms[idx1]
            atom_end   = u.atoms[idx2]

            dists = []

            for ts in u.trajectory:
                d = calc_bonds(
                    atom_start.position[np.newaxis, :],
                    atom_end.position[np.newaxis, :],
                    box=ts.dimensions
                )[0]
                dists.append(d)

            dists = np.array(dists)

            # Save distances only (KDE does not need time)
            np.savetxt(dat_file, np.column_stack((dists,)),
                       header="distance(nm)")
            print(f"  Saved â†’ {dat_file}")

        # =======================================================
        # 3. KDE COMPUTATION
        # =======================================================
        kde = gaussian_kde(dists, bw_method=bandwidth)

        x_grid = np.linspace(dists.min() * 0.9, dists.max() * 1.1, 500)
        y_vals = kde(x_grid)

        # Plot KDE curve
        plt.plot(x_grid, y_vals, lw=2, label=label)

    # Final plot formatting
    plt.xlabel("End-to-End Distance (nm)", fontsize=14)
    plt.ylabel("Probability Density", fontsize=14)
    plt.title("KDE of End-to-End Distance", fontsize=16)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()

    png_file = f"{output_prefix}.png"
    plt.savefig(png_file, dpi=300)
    plt.show()

    print(f"\nSaved KDE plot â†’ {png_file}")
```
